# Perform an outlier detection analysis on numerical variables (e.g., using the IQR method).

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20, 15))
numerical_features = housing.select_dtypes(include=np.number).columns
for i, col in enumerate(numerical_features):
    plt.subplot(5, 4, i + 1)
    sns.boxplot(y=housing[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
Q1=housing['Sale Price'].quantile(0.25)
Q3=housing['Sale Price'].quantile(0.75)

IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(Q1,Q3,IQR,lower_bound,upper_bound)
```

```python
housing['Sale Price']=housing['Sale Price'].clip(lower_bound,upper_bound)
sns.boxplot(housing['Sale Price'])
```

```python
#select numerical columns(exclude target variable(sale price))
Q=['No of Bedrooms','No of Bathrooms','Longitude','Flat Area (in Sqft)','Area of the House from Basement (in Sqft)','Living Area after Renovation (in Sqft)','Lot Area after Renovation (in Sqft)','Renovated Year','Basement Area (in Sqft)']
# Calculate Q1, Q3, IQR
Q1 = housing[Q].quantile(0.25)
Q3 = housing[Q].quantile(0.75)
IQR = Q3 - Q1
# Identify outliers using IQR rule
filter_out = ~((housing[Q] < (Q1 - 1.5 * IQR)) |
               (housing[Q] > (Q3 + 1.5 * IQR))).any(axis=1)

# Apply the filter
housing_filtered = housing[filter_out]

print("Original shape:", housing.shape)
print("After removing outliers (specific columns):", housing_filtered.shape)
```

```python
# Identify all categorical columns in the dataset.
categorical_cols = housing_filtered.select_dtypes(include='object').columns
print("Categorical Columns:",categorical_cols)
```

```python
#Apply appropriate encoding techniques:
# Use One-Hot Encoding for nominal categories.
# Use Label Encoding for ordinal categories, if applicable.
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder,StandardScaler,MinMaxScaler
label_enc=LabelEncoder()
housing_filtered['Condition of the House']=label_enc.fit_transform(housing_filtered['Condition of the House'])
housing_filtered
```

```python
housing_filtered=pd.get_dummies(housing_filtered,columns=['Waterfront View'],dtype=int,prefix='Waterfront View',drop_first=True)
housing_filtered
```

```python
# datetime-type column, not a categorical one — so instead of encoding it directly, the better approach is to extract meaningful features
#  Convert to datetime
housing_filtered['Date House was Sold'] = pd.to_datetime(housing_filtered['Date House was Sold'])
#Extract useful time-based features
housing_filtered['Sale_Year'] = housing_filtered['Date House was Sold'].dt.year
housing_filtered['Sale_Month'] = housing_filtered['Date House was Sold'].dt.month
housing_filtered['Sale_Weekday'] = housing_filtered['Date House was Sold'].dt.dayofweek  # 0=Monday, 6=Sunday
housing_filtered
```

```python
W=['No of Bedrooms',
 'No of Bathrooms',
 'Flat Area (in Sqft)',
 'Lot Area (in Sqft)',
 'No of Floors',
 'Condition of the House',
 'Overall Grade',
 'Area of the House from Basement (in Sqft)',
 'Basement Area (in Sqft)',
 'Age of House (in Years)',
 'Renovated Year',
 'Zipcode',
 'Latitude',
 'Longitude',
 'Living Area after Renovation (in Sqft)',
 'Lot Area after Renovation (in Sqft)',
 'Waterfront View_Yes',
 'Sale_Year',
 'Sale_Month',
 'Sale_Weekday']
min_scale=MinMaxScaler(feature_range=(0,1))
housing_filtered[W]=min_scale.fit_transform(housing_filtered[W])
housing_filtered
```

### Train-Test Split
### Set aside the SalePrice column as the target variable.
### Split the dataset into training (80%) and testing (20%) sets using the train_test_split function from sklearn.

```python
from sklearn.model_selection import train_test_split
# Step 1: Define features (x) and target (y)
x = housing_filtered.drop(columns=['Sale Price'])
y = housing_filtered['Sale Price']

# Step 2: Split the data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)  #)0.2 means 20  % data

print("shape of X_train:",x_train.shape)
print("shape of X_test:",x_test.shape)
print("shape of y_train:",y_train.shape)
print("shape of y_test:",y_test.shape)
```
