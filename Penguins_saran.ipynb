```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.shape
```

```python
penguins = penguins.dropna()
```

```python
penguins.isna().sum()
```

```python
penguins.columns
```

```python
# Used sns.scatterplot() to show the relationship between bill length and bill depth in the Penguins dataset.
sns.scatterplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm')
plt.title("Bill Length vs Bill Depth")
```

Highlights how bill length and depth vary across species by coloring points by species

```python
# Used sns.scatterplot() to plot bill length vs bill depth for penguins, with points colored by species using the hue parameter
sns.scatterplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
plt.title("Bill Length vs Bill Depth")
```

```python
# Used sns.pairplot() to visualize the relationships between all numeric features in the Penguins dataset, with points colored by species
sns.pairplot(data = penguins, hue = 'species')
```

```python
# Used sns.pairplot() to visualize the relationships between all numeric features in the Penguins dataset, with points colored by sex
sns.pairplot(data = penguins, hue = 'sex')
```

```python
#used sns.pairplot() to plot relationships between all numeric features in the Penguins dataset
sns.pairplot(data = penguins)
```

```python
# Used sns.boxplot() to compare flipper length across different penguin species
sns.boxplot(data = penguins, x = 'species', y = 'flipper_length_mm')
```

```python
# Selected only the numeric column from the dataset
penguins_num = penguins.select_dtypes(include=['int64','float64'])
penguins_num.head()
```

```python
# Calculated the correlation matrix for all numeric columns
corr_matrix = penguins_num.corr()
```

```python
# plotted a heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
# Used sns.violinplot() to show the distribution of flipper lengths for each penguin species
sns.violinplot(data = penguins, x = 'species', y = 'flipper_length_mm')
```

```python
# used sns.countplot() to show the count of penguins in each species
sns.countplot(data = penguins, x = 'species')
```

```python
#  used sns.stripplot() to display individual flipper length values for each penguin species
sns.stripplot(data = penguins, x = 'species', y = 'flipper_length_mm')
```

```python
# used sns.swarmplot() to show individual flipper length values for each penguin species
sns.swarmplot(data = penguins, x = 'species', y = 'flipper_length_mm')
```

```python
# used sns.kdeplot() to show the density distribution of penguins based on bill length and bill depth
sns.kdeplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm')
```

```python
# used sns.kdeplot() to plot the density of bill length vs bill depth, with separate contours for each penguin species using hue
sns.kdeplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
```
